### Iloc

iloc ialah cara cepat untuk slicing, cara define nya : .iloc[baris,kolom]

```python
#memilih banyak baris dan kolom
dataku.iloc[0:11]
```

```python
#memilih 1 kolom dan 1 baris9
dataku.iloc[:,3]
```

```python
#dalam pandas berlaku indexing minus
dataku.iloc[-5:-2,-9:-4]
```

```python
dataku.iloc[[1,2,3]]
```

```python
dataku.iloc[[1,2,3],[0,2,3]]
```

## loc

loc digunakan jika tidak ingin menggunakan index nya dalam slicing

```python
data_baru = pd.read_csv('csv/kapal_titanic.csv', index_col = 'embarked')
```

```python
data_baru
```

```python
data_baru.loc['S']
```

```python
data_baru.loc['S','sex']
```

```python
data_baru.loc[['S','C'],['sex','fare']]
```
