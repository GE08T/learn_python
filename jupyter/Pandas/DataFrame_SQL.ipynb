## menggabungkan dengan unique key persis seperti sql 

```python
import pandas as pd
```

```python
satu = pd.DataFrame({
    'key':['k0','k1','k2','k3','k4'],
    'one':[1,2,3,4,5],
    'two':[6,7,8,9,10]
})
```

```python
dua = pd.DataFrame({
    'key':['k0','k1','k2','k3','k4'],
    'three':[1,2,3,4,5],
    'four':[6,7,8,9,10]
})
```

```python
satu
```

```python
dua
```

## Merge

```python
#menggabungkan satu dan dua dengan merge dengan key didalam kolom
baru = pd.merge(satu,dua, how = 'inner', on = 'key')
```

```python
baru
```

```python
satu = pd.DataFrame({
    'key1':['K0','K0','K1','K2'],
    'key2':['K0','K1','K0','K1'],
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
})

dua = pd.DataFrame({
    'key1':['K0','K1','K1','K2'],
    'key2':['K0','K0','K0','K0'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3'],
})
```

```python
satu
```

```python
dua
```

```python
#membuat merge dengan left
#satu akan menjadi patokan
#jika dua false tidak akan tampil
baru2 = pd.merge(satu, dua, how = 'left', on = ['key1','key2'])
```

```python
baru2
```

```python
#membuat merge dengan right
#dua akan menjadi patokan
#jika satu false tidak akan tampil
baru2 = pd.merge(satu, dua, how = 'right', on = ['key1','key2'])
```

```python
baru2
```

```python
#membuat merge dengan outer
#tidak ada patokan
#jika satu atau dua false tidak akan tampil
baru2 = pd.merge(satu, dua, how = 'outer', on = ['key1','key2'])
```

```python
baru2
```

```python
#biasanya yang sering dipakai adalah inner
```

## Join

```python
a = pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2']},
    index = ['K0','K1','K2'])

b = pd.DataFrame({
    'C':['C0','C1','C2'],
    'D':['D0','D1','D2']},
    index = ['K0','K2','K3'])
```

```python
a
```

```python
b
```

```python
#menggabungkan satu dan dua dengan merge dengan key didalam index
#default nya how di join adalah left
a.join(b)
```

```python
#akan digabungkan jika ke 2 frame benar ada data keynya
a.join(b, how = 'inner')
```
