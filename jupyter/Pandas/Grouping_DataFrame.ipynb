## Menggabungkan Data Frame 

```python
import pandas as pd 
```

```python
satu = pd.DataFrame({
                    'A':['a0','a1','a2','a3','a4','a5'],
                    'B':['b0','b1','b2','b3','b4','b5'],
                    'C':['c0','c1','c2','c3','c4','c5'],
                    'D':['d0','d1','d2','d3','d4','d5'],
                    'E':['e0','e1','e2','e3','e4','e5'],
                    }, index = [0,1,2,3,4,5])
```

```python
dua = pd.DataFrame({
                    'A':['a6','a7','a8','a9','a10','a11'],
                    'B':['b6','b7','b8','b9','b10','b11'],
                    'C':['c6','c7','c8','c9','c10','c11'],
                    'D':['d6','d7','d8','d9','d10','d11'],
                    'E':['e6','e7','e8','e9','e10','e11'],
                    }, index = [6,7,8,9,10,11])
```

```python
tiga = pd.DataFrame({
                    'A':['a12','a13','a14','a15','a16','a17'],
                    'B':['b12','b13','b14','b15','b16','b17'],
                    'C':['c12','c13','c14','c15','c16','c17'],
                    'D':['d12','d13','d14','d15','d16','d17'],
                    'E':['e12','e13','e14','e15','e16','e17'],
                    }, index = [12,13,14,15,16,17])
```

```python
satu
```

```python
dua
```

```python
tiga
```

```python
#membuat group dengan concate
pd.concat([satu,dua,tiga])
```

```python
#membuat group dengan concate dengan merging colom 
pd.concat([satu,dua,tiga], axis = 1)
```

```python
bisnis = pd.DataFrame({
    'Perusahaan':['Pertamina','Pertamina','Telkom','BRI','Telkom','KFC','KFC','Indosat'],
    'Karyawan':['Joko','Budi','Setiawan','Susi','Bambang','Iriana','Burhan','Intan'],
    'Usia': [29,34,23,26,25,43,32,41]
})
```

```python
bisnis
```

```python
#menggunakan group by 
bisnis.groupbyby('Perusahaan')
```

```python
#menjumlahkan type data numerik sesuai group perusahaan
bisnis.groupby('Perusahaan').sum()
```

```python
#menghitung berapa banyak data per perusahaan
bisnis.groupby('Perusahaan').count()
```

```python
#menghitung rata rata 
bisnis.groupby('Perusahaan').mean()
```

```python
#diurutkan dari yang paling kecil dari Perusahaan
bisnis.groupby('Perusahaan').min()
```

```python
#diurutkan dari yang paling kecil dari Perusahaan
bisnis.groupby('Perusahaan').max()
```

```python
#perlu diperhatikan ini akan mengambil semua data tidak salah satu
#jika ingin salah satu gunakan dot notation
bisnis.groupby('Perusahaan').Usia.max()
```

```python
#juga dapat menggunakan fungsi describe
bisnis.groupby('Perusahaan').describe()
```
