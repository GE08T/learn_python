```python
import pandas as pd
```

```python
dataku = pd.read_csv('datacsv.csv')
```

```python
#membuat histogram dengan pandas
dataku['age'].hist()
```

```python
#kita juga dapat mengatur bins = batang
dataku.age.hist(bins = 5)
```

```python
#dapat di perbagus lagi dengan
dataku.age.plot.hist(bins = 5)
```

```python
#atau dengan seaborn
import seaborn as sns
sns.set()
```

```python
dataku.age.plot.hist(bins = 30)
```

```python
#mengganti transparansi menggunakan alpha
#default nya 1
dataku.age.plot.hist(bins = 30, alpha = 0.5)
```

```python
import numpy as np
```

```python
np.random.seed(100)
```

```python
angka = []
for i in range(0,100):
    angka.append(i)
```

```python
baru = pd.DataFrame(np.random.rand(100,5), angka, ['A','B','C','D','E'])
```

```python
baru.iloc[0:11]
```

```python
#membuat area plot dengan pandas
#atau luas permukaan 
baru.plot.area(y = 'B',alpha = 0.5)
```

```python
baru2 = pd.DataFrame(np.random.rand(4,5), [0,1,2,3], ['A','B','C','D','E'])
```

```python
baru2
```

```python
#membuat grafik bar
#stacked default false
baru2.plot.bar(stacked = True)
```

```python
#mengambil hanya kolom e
baru2['E'].plot.bar()
```

```python
#membuat diagram scatter = diagram tebar
#dengan sumbu x adalah a dan y adalah b
#c = color
baru.plot.scatter(x = 'A', y = 'B', c = 'E', cmap = 'ocean')
```

```python
#membuat scatter diagram dengan scalar 
#jika terdapat titik yang lebih besar 
#maka nilai titik di D nya lebih besar
baru.plot.scatter(x = 'A', y = 'B', s = baru.D*100, c = 'E', cmap = 'ocean')
```

```python
#membuat diagram box
#dalam diagram dapat melihat perbedaan nilai median
#nilai kuartil 1 = 25%, nilai kuartil 2 = 75%
baru.plot.box()
```

```python
#membuat diagram line 
#require params = y
#lw = line weight
#figsize = panjang lebar line plotnya,
#20 = lebar, 5 = panjang 
baru.plot.line(y = 'E', lw = 0.5, figsize = (20,5))
```

```python
#diagram hexbin / scatter dengan bentuk hexagonal
baru.plot.hexbin(x = 'A', y = 'B', gridsize = 15)
```
