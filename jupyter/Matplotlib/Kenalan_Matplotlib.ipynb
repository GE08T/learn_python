### Pengetahuan Dasar Grafik

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#untuk membuat mentahan grafik
plt.plot()
```

```python
#atau menggunakan 
#perbedaan nya hanya menghilangkan output 
plt.plot();
plt.show()
```

```python
#mengganti nilai sumbu y (tegak) dan x (miring)
plt.plot([10,20,30,40,50],[1,2,3,4,5])
```

```python
x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
#atau dengan mengganti seperti ini
plt.plot(x,y)
#menambahkan judul
plt.title('Judul Bro')
#menambahkan label sumbu x
plt.xlabel('Sumbu x')
#menambahkan label sumbu y
plt.ylabel('Sumbu y')
plt.show()
```

```python
#membuat 2 grafis dengan subplot
plt.subplot(121)
plt.plot(x,y, color = 'cyan')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

plt.subplot(122)
plt.plot(x,y, color = 'maroon')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')

#untuk menyesuaikan lebar 
plt.tight_layout()
plt.show()
```
